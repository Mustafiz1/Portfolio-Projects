```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
import seaborn as sns
plt.style.use('ggplot')
from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
## missing data

df.isnull().sum()
```

```python
df = df.dropna(axis = 0)
df.isnull().sum()
```

```python
df.shape
```

```python
# percentage of missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types

df.dtypes
```

```python
# change budget column type

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['year_corrected'] = df['released'].astype(str).str.split(',').str[1].str[:5]
df['year_corrected']
```

```python
df.head()
```

```python

df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
## Display all the rows

pd.set_option('display.max_rows', None)
```

```python
# drop duplicates

df.drop_duplicates().head()
```

```python
df.shape
```

```python
# correlation 

plt.scatter(x = df['budget'], y= df['gross'])
plt.title('budget Vs Gross earnings')
plt.xlabel('budget')
plt.ylabel('Gross earnings')

plt.show()
```

```python
df.head()
```

```python
# plot the graph in sns

sns.regplot(x ='budget', y= 'gross', 
            data = df, 
            scatter_kws ={'color' :"red"}, line_kws={'color':'green'})
```

```python
df.corr(method = 'pearson')  # pearson, kendrall, spearman
```

```python
corr_matrix = df.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation of features')
plt.xlabel('features')
plt.ylabel('features')

plt.show()
```

```python
# Transform the company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes
            
df_numerized.head()           
```

```python
df.head()
```

```python
sns.heatmap(df_numerized, annot = True)
plt.title('Correlation of features')
plt.xlabel('features')
plt.ylabel('features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs> 0.5]

high_corr.head()
```
